```python
def analyze_raw_moisture_temperature(excel_file="all_data_with_delta.xlsx"):

    df = pd.read_excel(excel_file)

    sns.set(style="whitegrid", context="talk")

    # Sort so the bar charts go in ascending SoakingTime order
    df = df.sort_values(["Trial", "Bag", "SoakingTime", "Sample"])

    trials = df["Trial"].unique()

    for trial in trials:
        sub = df[df["Trial"] == trial]

        g1 = sns.catplot(data=sub, x="SoakingTime", y="MC%", hue="Sample", col="Bag", kind="bar", palette="Set2", height=5, aspect=1)
        g1.fig.suptitle(f"Trial {trial} - Moisture Content (MC%)", y=1.02)
        g1.set_axis_labels("Soaking Time (hours) [0=Initial]", "MC%")
        g1._legend.set_title("Sample")

        plt.show()


if __name__ == "__main__":
    analyze_raw_moisture_temperature("all_data_with_delta.xlsx")
```

```python
# Load the Excel file
df = pd.read_excel("all_large_data.xlsx")


# Function to predict moisture at specific soaking times
def predict_moisture(df, trial, predict_times):
    df_trial = df[df["Trial"] == trial]
    df_avg = df_trial.groupby("SoakingTime")["MC%"].mean().reset_index()

    # Interpolation function
    interp_func = interp1d(df_avg["SoakingTime"], df_avg["MC%"], kind="linear", fill_value="extrapolate")

    # Predict moisture content for provided times
    predictions = {time: float(interp_func(time)) for time in predict_times}

    return predictions


# Trial 1: Moisture at 4 hours
trial1_predictions = predict_moisture(df, trial=1, predict_times=[4])
print("Trial 1 moisture prediction at 4 hours:", trial1_predictions[4])

# Trial 2: Moisture at 11 hours 15 minutes (11.25 hours)
trial2_predictions = predict_moisture(df, trial=2, predict_times=[11.25])
print("Trial 2 moisture prediction at 11 hours 15 minutes:", trial2_predictions[11.25])

# Trial 3: Moisture at 25 hours 40 minutes (25.6667 hours) and 28 hours 20 minutes (28.3333 hours)
trial3_predictions = predict_moisture(df, trial=3, predict_times=[25.6667, 28.3333])
print("Trial 3 moisture prediction at 25 hours 40 minutes:", trial3_predictions[25.6667])
print("Trial 3 moisture prediction at 28 hours 20 minutes:", trial3_predictions[28.3333])
```

```python
# Define a logarithmic function for curve fitting
def log_func(t, a, b, c):
    return a * np.log(t + b) + c


# Function to predict moisture at specific soaking times using logarithmic model
def predict_moisture_log(df, trial, predict_times):
    df_trial = df[df["Trial"] == trial]
    df_avg = df_trial.groupby("SoakingTime")["MC%"].mean().reset_index()

    # Fit the logarithmic model
    params, _ = curve_fit(log_func, df_avg["SoakingTime"], df_avg["MC%"], maxfev=5000)

    # Predict moisture content for provided times
    predictions = {time: float(log_func(time, *params)) for time in predict_times}

    return predictions


# Extended predictions for longer soaking times
extended_times = [33, 36.75, 50, 51.5, 53.0833]

# Predicting for Trial 3 (assuming it's the trial with longest available data)
extended_predictions = predict_moisture_log(df, trial=3, predict_times=extended_times)
for time, moisture in extended_predictions.items():
    hours = int(time)
    minutes = int((time - hours) * 60)
    print(f"Trial 3 moisture prediction at {hours} hours {minutes} minutes: {moisture:.2f}%")
```
